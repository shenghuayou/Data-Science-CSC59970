```python
r2 = requests.get('http://www.gapminder.org/documentation/documentation/gapdata001%20v14.xlsx')
income = pd.read_excel(io.BytesIO(r2.content),sheetname=1)
income = income.loc[:,['Area', 'Year', 'GDP per capita - with interpolations']]
income.head()
```

#### Problem 2(b)

Graphically display the distribution of income per person across all countries in the world for any given year (e.g. 2000).  What kind of plot would be best? 

```python
m_year=2000
new_income=income[income.Year==m_year]
my_list=[]
my_list2=[]

for i in range(0,len(new_income)):
    my_list.append(new_income.iloc[i].values[2])

for i in range(0,len(new_income)):
    my_list2.append(new_income.iloc[i].values[0])    

my_list_GDP=np.array(my_list)
my_list_Area=np.array(my_list2)

plt.figure(num=None, figsize=(40,15))
plt.xticks(range(len(my_list_Area)), [i for i in my_list_Area],rotation='vertical')
plt.bar(range(0,len(new_income)),my_list_GDP)
plt.title('2000 income per person')
plt.show()
```

#### Problem 2(c)

Write a function to merge the `countries` and `income` data sets for any given year. 

```python
"""
Function
--------
mergeByYear

Return a merged DataFrame containing the income, 
country name and region for a given year. 

Parameters
----------
year : int
    The year of interest

Returns
-------
a DataFrame
   A pandas DataFrame with three columns titled 
   'Country', 'Region', and 'Income'. 

Example
-------
>>> mergeByYear(2010)
"""
#your code here
def mergeByYear(year):
    my_income=income[income.Year==year]
    my_income = my_income.rename(columns={'Area': 'Country', 'GDP per capita - with interpolations': 'Income'})
    my_income = my_income.loc[:,['Country', 'Income']]  
    merge_data = pd.merge(countries, my_income, on=['Country'], how='inner')
    return merge_data

mergeByYear(2010).head()
```

#### Problem 2(d) 

Use exploratory data analysis tools such as histograms and boxplots to explore the distribution of the income per person by region data set from 2(c) for a given year. Describe how these change through the recent years?

**Hint**: Use a `for` loop to consider multiple years.  

```python
begin=2000
end=2005
def drawfig(year):
    my_data = mergeByYear(year)
    all_regions = ['AFRICA', 'ASIA', 'EUROPE', 'NORTH AMERICA', 'OCEANIA', 'SOUTH AMERICA']
    i=1
    my_value_list=[]
    for r in all_regions:
        my_region_data= my_data[my_data.Region==r]
        my_value=my_data[my_data.Region==r].loc[:,['Income']].values[:,0]
        my_value_list.append(my_value)
        a = np.random.normal(i, 0.1, size=len(my_value))
        plt.scatter(a,my_value,alpha=0.6)
        i=i+1

    plt.boxplot(my_value_list)
    plt.xticks(range(1,len(all_regions)+1), all_regions,rotation='vertical')
    plt.xlabel('regions')
    plt.ylabel('income')
    plt.title('income per person by regions in '+str(year))
    plt.show()
    
for y in range(begin,end):
    drawfig(y)
```

## Problem 3: Is the average of polls better than just one poll?

#### Problem 3(a)

Most undecided voters vote for one of the two candidates at the election. Therefore, the reported percentages underestimate the final value of both candidates. However, if we assume the undecided will split evenly, then the observed difference should be an unbiased estimate of the final difference. 

Add a new column to the `election` DataFrame containg the difference between Obama and Romeny called `Diff`. 

```python
r = requests.get('http://elections.huffingtonpost.com/pollster/2012-general-election-romney-vs-obama.csv')
election=pd.read_csv(io.BytesIO(r.content))
election.head()
```

```python
election['Diff']=election['Obama']-election['Romney']
election['Diff']=election['Diff']/100
election.head()
```

#### Problem 3(b)

Make a plot of the differences for the week before the election (e.g. 5 days) where the days are on the x-axis and the differences are on the y-axis.  Add a horizontal line showing 3.9%: the difference between Obama and Romney on election day.

```python
my_days = election[(election['End Date'] > '2012-10-26') & (election['End Date'] <= '2012-10-31')]
plt.figure(figsize=(10,5))
plt.scatter(pd.to_datetime(my_days['End Date'].values),my_days.Diff.values)
plt.axhline(y=0.039)
plt.xlabel("days")
plt.ylabel('Diff')
plt.title('the week before the election')
plt.show()
```

#### Problem 3(c) 

Make a plot showing the differences by pollster where the pollsters are on the x-axis and the differences on the y-axis. 

```python
x_range=range(len(my_days.Pollster.values))
plt.figure(figsize=(14,7))
plt.scatter(x_range,my_days.Diff.values)
plt.xticks(x_range, my_days.Pollster.values)
plt.xticks(rotation=90)
plt.xlabel('Pollster')
plt.ylabel('Diff')
plt.title('Pollster and Diff')
plt.grid(True)
plt.show()
```

Is the *across poll* difference larger than the *between pollster* difference? 

Answer: Yes,across poll is larger.

#### Problem 3(d)

Take the average for each pollster and then compute the average of that. Given this difference how confident would you have been of an Obama victory?

**Hint**: Compute an estimate of the SE of this average based exclusively on the observed data. 

```python
pollster_mean = election.groupby(election.Pollster).mean()
my_mean=np.mean(pollster_mean['Diff'].values)
my_std=np.std(pollster_mean['Diff'].values)
print('mean:',my_mean)
print('std:',my_std)
```

Answer: since SE and diff mean are small, we can't be sure about victory

#### Problem 3(e)

Show the difference against time and see if you can detect a trend towards the end. Use this trend to see if it improves the final estimate.

```python
all_2012_days=election[election['End Date']>='2012-01-01']
all_2012_days = all_2012_days.groupby('End Date').mean()
plt.figure(figsize=(16,8))
plt.plot(pd.to_datetime(all_2012_days.index).values, all_2012_days['Diff'].values)
plt.xlabel('month')
plt.ylabel('Diff')
plt.title('Diff through the year')
plt.axhline(y=0,c='g')
plt.show()
```

Answer: at the end of election, the Diff is close. Before Oct2012, obama has advanages. So i can't detect the trend at the end.
